```python
import math
import matplotlib.pyplot as plt
def normal(x, mu, sigma): # 变量， 均值， 方差
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    return p * np.exp(-0.5 / sigma**2 * (x - mu)**2)

x = np.arange(-6, 6, 0.01) ## 每0.01 为刻度
params = [(0, 1), (0, 2), (3, 1)] # 均值， 方差
test = [normal(x, mu, sigma) for mu, sigma in params]
for index, i in enumerate(test):
    plt.plot(x, np.array(i), label=index)
plt.legend()    
```

### 从零开始

**生成数据集**

```python
def synthetic_data(w, b, num):
    x = torch.normal(0, 1, size=(num, len(w))) # len(w) 特征数量
    y = torch.matmul(x, w) + b  # 矩阵乘法， 加上偏移量 1000， 2， 2 ，1 
    y += torch.normal(0, 0.01, y.shape) # 加上噪声
    return x, y.reshape((-1, 1))
```

```python
true_w = torch.tensor([2, -3.4]) # 真实的特征
true_b = 4.2  # 真实的偏移量
fatures, labes = synthetic_data(true_w, true_b, 1000)
```

### 处理数据集

**处理小批量数据**

```python
import random
def data_iter(batch_size, features, labels):
    num = len(features)
    indices = list(range(num))
    # 随机切分
    random.shuffle(indices) # 打乱下标
    for i in range(0, num, batch_size):
        batch_indices = torch.tensor(indices[i: min(i + batch_size, num)]) 
        yield features[batch_indices], labels[batch_indices] # 迭代器迭代
```

```python
for X, y in data_iter(10, fatures, labes):
    print(X, '\n', y)
```

**初始化模型参数**

```python
# 权重按正态分布，初始化， offset设置为0
w = torch.normal(0, 0.01, size=(2,1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)
```

**定义模型**

```python
def linreg(X, w, b):
    """线性回归模型"""
    return torch.matmul(X, w) + b
```

**定义损失函数**

```python
def squared_loss(y_hat, y):
#     损失函数
    return (y_hat - y.reshape(y_hat.shape)) ** 2 / 2
```

**定义优化算法**

```python
# 学习速率控制批量的大小
def sgd(params, lr, batch_size):
    with torch.no_grad(): # 更新的时候不需要求，梯度，只有训练的时候才需要梯度
        for param in params:
            param  -= lr * param.grad / batch_size
            param.grad.zero_()
```

**训练模型**

```python
lr = 0.03  # 学习速率
num_opochs = 3 # 迭代周期
net = linreg # 模型
less = squared_loss # 损失函数
```

```python
for epoch in range(num_opochs):
    for X, y in data_iter(batch_size=10, features=fatures, labels=labes):
        l = less(net(X, w, b), y)
        # 因为l形状的（batch_size, 1)不是一个标量， 所以需要计算标量 sum
        l.sum().backward() # 通过损失函数来 计算梯度。
        sgd([w, b], lr, 10) # 更新参数
    with torch.no_grad():
        train_l = less(net(fatures, w, b), labes)
        print(f'epoch {epoch + 1}, less {float(train_l.mean()):f}')
```

**对比训练值和真实值**

```python
print(f'w的估计误差: {true_w - w.reshape(true_w.shape)}')
print(f'b的估计误差: {true_b - b}' )
```

### 线性回归的简单实现

```python
import numpy as np
import torch
from torch.utils import data
```

```python
true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = synthetic_data(true_w, true_b, 1000)
```

**读取数据**

```python
## 迭代读取数据
def load_array(data_arrays, batch_size, is_train=True):
    dataset = data.TensorDataset(*data_arrays) ## 可以读取多个数组
    return data.DataLoader(dataset, batch_size, shuffle=is_train)
```

```python
batch_size = 10
data_itor = load_array((features, labels), batch_size)
next(iter(data_itor))
```

**定义模型**

```python
from torch import nn ## mm 是神经网络的缩写
net = nn.Sequential(nn.Linear(2, 1)) ## 两个参数， 1， 输入特征数量 2， 输入特征的形状
```

全连接层：每一个输入都通过矩阵-向量乘法得到它的每个输出

**初始化模型参数**

```python
# net[0] 表示选择网络中的第一个图层, 使用weight.data 和 bias.data 访问参数
net[0].weight.data.normal_(0, 0.01) 
net[0].bias.data.fill_(0)
```

**定义损失函数**

```python
loss = nn.MSELoss()  # 计算均方误差
```

**定义优化算法**

```python
trainer = torch.optim.SGD(net.parameters(), lr=0.03)
```

**训练**

```python
num_epochs = 3
for epoch in range(num_epochs):
    for X, y in data_itor: 
        l = loss(net(X) ,y) 
        trainer.zero_grad() 
        l.backward()  # 计算梯度
        trainer.step() # 更新模型
    l = loss(net(features), labels) # 计算每一个 损失
    print(f'epoch {epoch + 1}, loss {l:f}')
```

## softmax回归从0开始实现

```python
import torch
from IPython import display 
from d2l import torch as d2l
# 随机读取256张图片
batch_size = 256
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
```

```python
num_inputs = 784 # 将图片展平
num_outputs = 10

W = torch.normal(0, 0.01, size=(num_inputs, num_outputs), requires_grad=True) 
b = torch.zeros(num_outputs, requires_grad=True)# 每个输出 都有一个 偏移
```

**实现softmax**函数

```python
def softmax(X):
    X_exp = torch.exp(X)
    partition = X_exp.sum(1, keepdim=True) # 求和， 处以所有和
    return X_exp / partition # 这里应用了广播机制
```

```python
X = torch.normal(0, 1, (2, 5))
X_prob = softmax(X)
X_prob, X_prob.sum(1)
```

**实现softmax回归模型**

```python
X.shape, W.shape
```

```python
# reshape 将向量 平铺
def net(X):
    return softmax(torch.matmul(X.reshape((-1, W.shape[0])), W) + b)
```

**损失函数**

```python
#  one 第一个是正确的， 第二个样本中， 第3个是正确的概率
y = torch.tensor([0, 2])  # 正确概率的下标
# 两个样本在 3 个类别的预测概率
y_hat = torch.tensor([[0.1, 0.3, 0.6], [0.3, 0.2, 0.5]])  # 数据样本， 2 个样本， 3个类别的预测概率，
y_hat[[0, 1], y] # 第几个样本，
```

```python
def cross_entropy(y_hat, y): # 交叉熵损失函数。 取出真实标号的 对应的预测值， y是正确预测值的 下标
    return - torch.log(y_hat[range(len(y_hat)), y])
cross_entropy(y_hat, y) # 样本0， 样本1 的损失。
```

**预测值与真实值比较**

```python
def accuracy(y_hat, y): #@save 
    """计算预测正确的数量"""
    if len(y_hat.shape) > 1 and y_hat.shape[1] > 1:
        y_hat = y_hat.argmax(axis=1)
    cmp = y_hat.type(y.dtype) == y 
    return float(cmp.type(y.dtype).sum())
```

```python
accuracy(y_hat, y) / len(y)
```

**评估任意模型精度**

```python
def evaluate_accuracy(net, data_iter): #@save """计算在指定数据集上模型的精度"""
    if isinstance(net, torch.nn.Module):
        net.eval() # 将模型设置为评估模式
    metric = Accumulator(2) # 正确预测数、预测总数 
    with torch.no_grad():
        for X, y in data_iter: 
            metric.add(accuracy(net(X), y), y.numel())
    return metric[0] / metric[1]
```

**实用程序类**

```python
class Accumulator: #@save 
    """在n个变量上累加""" 
    def __init__(self, n):
        self.data = [0.0] * n 
    def add(self, *args):
        self.data = [a + float(b) for a, b in zip(self.data, args)] 
    def reset(self):
        self.data = [0.0] * len(self.data)
    def __getitem__(self, idx): 
        return self.data[idx]
```

```python
evaluate_accuracy(net, test_iter)
```

### 训练

```python
def train_epoch_ch3(net, train_iter, loss, updater): #@save 
    """训练模型一个迭代周期(定义⻅第3章)"""
    # 将模型设置为训练模式
    if isinstance(net, torch.nn.Module):
        net.train()
    # 训练损失总和、训练准确度总和、样本数 
    metric = Accumulator(3)
    for X, y in train_iter:
        # 计算梯度并更新参数
        y_hat = net(X)
        l = loss(y_hat, y)
        if isinstance(updater, torch.optim.Optimizer):
            # 使用PyTorch内置的优化器和损失函数 
            updater.zero_grad() 
            l.mean().backward() 
            updater.step()
        else:
            # 使用定制的优化器和损失函数 
            l.sum().backward() 
            updater(X.shape[0])
        metric.add(float(l.sum()), accuracy(y_hat, y), y.numel()) 
    # 返回训练损失和训练精度
    return metric[0] / metric[2], metric[1] / metric[2]
```

**绘图**

```python
class Animator:  #@save
    """在动画中绘制数据"""
    def __init__(self, xlabel=None, ylabel=None, legend=None, xlim=None,
                 ylim=None, xscale='linear', yscale='linear',
                 fmts=('-', 'm--', 'g-.', 'r:'), nrows=1, ncols=1,
                 figsize=(3.5, 2.5)):
        # 增量地绘制多条线
        if legend is None:
            legend = []
        d2l.use_svg_display()
        self.fig, self.axes = d2l.plt.subplots(nrows, ncols, figsize=figsize)
        if nrows * ncols == 1:
            self.axes = [self.axes, ]
        # 使用lambda函数捕获参数
        self.config_axes = lambda: d2l.set_axes(
            self.axes[0], xlabel, ylabel, xlim, ylim, xscale, yscale, legend)
        self.X, self.Y, self.fmts = None, None, fmts

    def add(self, x, y):
        # 向图表中添加多个数据点
        if not hasattr(y, "__len__"):
            y = [y]
        n = len(y)
        if not hasattr(x, "__len__"):
            x = [x] * n
        if not self.X:
            self.X = [[] for _ in range(n)]
        if not self.Y:
            self.Y = [[] for _ in range(n)]
        for i, (a, b) in enumerate(zip(x, y)):
            if a is not None and b is not None:
                self.X[i].append(a)
                self.Y[i].append(b)
        self.axes[0].cla()
        for x, y, fmt in zip(self.X, self.Y, self.fmts):
            self.axes[0].plot(x, y, fmt)
        self.config_axes()
        display.display(self.fig)
        display.clear_output(wait=True)
```

**可视化**

```python
def train_ch3(net, train_iter, test_iter, loss, num_epochs, updater):  #@save
    """训练模型（定义见第3章）"""
    animator = Animator(xlabel='epoch', xlim=[1, num_epochs], ylim=[0.3, 0.9],
                        legend=['train loss', 'train acc', 'test acc'])
    for epoch in range(num_epochs):
        train_metrics = train_epoch_ch3(net, train_iter, loss, updater)
        test_acc = evaluate_accuracy(net, test_iter)
        animator.add(epoch + 1, train_metrics + (test_acc,))
    train_loss, train_acc = train_metrics
    assert train_loss < 0.5, train_loss
    assert train_acc <= 1 and train_acc > 0.7, train_acc
    assert test_acc <= 1 and test_acc > 0.7, test_acc
```

**学习率**

```python
lr = 0.1

def updater(batch_size):
    return d2l.sgd([W, b], lr, batch_size)
```

```python
num_epochs = 10
train_ch3(net, train_iter, test_iter, cross_entropy, num_epochs, updater)
```

### softmax简单实现

```python
import torch
from torch import nn
from d2l import torch as d2l
```

```python
batch_size = 256
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
```

**!.初始化模型参数**

```python
# PyTorch不会隐式地调整输入的形状。因此，
# 我们在线性层前定义了展平层（flatten），来调整网络输入的形状
net = nn.Sequential(nn.Flatten(), nn.Linear(784, 10))

# 初始化权重
def init_weights(m):
    if type(m) == nn.Linear:
        nn.init.normal_(m.weight, std=0.01)

net.apply(init_weights);
```

**2.损失函数**

```python
loss = nn.CrossEntropyLoss(reduction='none')
```

**3.优化算法**

```python
trainer = torch.optim.SGD(net.parameters(), lr=0.1)
```

**4.训练模型**

```python
num_epochs = 10
d2l.train_ch3(net, train_iter, test_iter, loss, num_epochs, trainer)
```
