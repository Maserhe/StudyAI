### 正态分布

![截屏2022-06-11 03.26.05](https://ghproxy.com/https://github.com/Maserhe/PIc-Bed/blob/master/typora/202206110326587.png)

```python
import math
import matplotlib.pyplot as plt
def normal(x, mu, sigma): # 变量， 均值， 方差
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    return p * np.exp(-0.5 / sigma**2 * (x - mu)**2)

x = np.arange(-6, 6, 0.01) ## 每0.01 为刻度
params = [(0, 1), (0, 2), (3, 1)] # 均值， 方差
test = [normal(x, mu, sigma) for mu, sigma in params]
for index, i in enumerate(test):
    plt.plot(x, np.array(i), label=index)
plt.legend()    
```

### 从零开始

**生成数据集**

```python
def synthetic_data(w, b, num):
    x = torch.normal(0, 1, size=(num, len(w))) # len(w) 特征数量
    y = torch.matmul(x, w) + b  # 矩阵乘法， 加上偏移量 1000， 2， 2 ，1 
    y += torch.normal(0, 0.01, y.shape) # 加上噪声
    return x, y.reshape((-1, 1))
```

```python
true_w = torch.tensor([2, -3.4]) # 真实的特征
true_b = 4.2  # 真实的偏移量
fatures, labes = synthetic_data(true_w, true_b, 1000)
```

### 处理数据集

**处理小批量数据**

```python
import random
def data_iter(batch_size, features, labels):
    num = len(features)
    indices = list(range(num))
    # 随机切分
    random.shuffle(indices) # 打乱下标
    for i in range(0, num, batch_size):
        batch_indices = torch.tensor(indices[i: min(i + batch_size, num)]) 
        yield features[batch_indices], labels[batch_indices] # 迭代器迭代
```

```python
for X, y in data_iter(10, fatures, labes):
    print(X, '\n', y)
```

**初始化模型参数**

```python
# 权重按正态分布，初始化， offset设置为0
w = torch.normal(0, 0.01, size=(2,1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)
```

**定义模型**

```python
def linreg(X, w, b):
    """线性回归模型"""
    return torch.matmul(X, w) + b
```

**定义损失函数**

```python
def squared_loss(y_hat, y):
#     损失函数
    return (y_hat - y.reshape(y_hat.shape)) ** 2 / 2
```

**定义优化算法**

```python
# 学习速率控制批量的大小
def sgd(params, lr, batch_size):
    with torch.no_grad():
        for param in params:
            param  -= lr * param.grad / batch_size
            param.grad.zero_()
```

**训练模型**

```python
lr = 0.03  # 学习速率
num_opochs = 3 # 迭代周期
net = linreg # 模型
less = squared_loss # 损失函数
```

```python
for epoch in range(num_opochs):
    for X, y in data_iter(batch_size=10, features=fatures, labels=labes):
        l = less(net(X, w, b), y)
        # 因为l形状的（batch_size, 1)不是一个标量， 所以需要计算标量 sum
        l.sum().backward()
        sgd([w, b], lr, 10) # 更新参数
    with torch.no_grad():
        train_l = less(net(fatures, w, b), labes)
        print(f'epoch {epoch + 1}, less {float(train_l.mean()):f}')
```

**对比训练值和真实值**

```python
print(f'w的估计误差: {true_w - w.reshape(true_w.shape)}')
print(f'b的估计误差: {true_b - b}')
```

### 线性回归的简单实现

```python
import numpy as np
import torch
from torch.utils import data
```

```python
true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = synthetic_data(true_w, true_b, 1000)
```

**读取数据**

```python
## 迭代读取数据
def load_array(data_arrays, batch_size, is_train=True):
    dataset = data.TensorDataset(*data_arrays) ## 可以读取多个数组
    return data.DataLoader(dataset, batch_size, shuffle=is_train)
```

```python
batch_size = 10
data_itor = load_array((features, labels), batch_size)
next(iter(data_itor))
```

**定义模型**

```python
from torch import nn ## mm 是神经网络的缩写
net = nn.Sequential(nn.Linear(2, 1)) ## 两个参数， 1， 输入特征数量 2， 输入特征的形状
```

全连接层：每一个输入都通过矩阵-向量乘法得到它的每个输出

**初始化模型参数**

```python
# net[0] 表示选择网络中的第一个图层, 使用weight.data 和 bias.data 访问参数
net[0].weight.data.normal_(0, 0.01) 
net[0].bias.data.fill_(0)
```

**定义损失函数**

```python
loss = nn.MSELoss()  # 计算均方误差
```

**定义优化算法**

```python
trainer = torch.optim.SGD(net.parameters(), lr=0.03)
```

**训练**

```python
num_epochs = 3
for epoch in range(num_epochs):
    for X, y in data_itor: 
        l = loss(net(X) ,y) 
        trainer.zero_grad() 
        l.backward() 
        trainer.step()
    l = loss(net(features), labels) # 计算每一个 损失
    print(f'epoch {epoch + 1}, loss {l:f}')
```
