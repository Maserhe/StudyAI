**创建白色网格图**

```python
%matplotlib inline
plt.style.use('seaborn-whitegrid')
fig = plt.figure() # 图形
ax = plt.axes() # 坐标轴

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
```

### plot中的参数
- color 颜色
- linestyle  线的风格 solid 实线、 dashed虚线、 dashdot点华线、 dotted实点线

```python
plt.plot(x, x + 4, linestyle='solid')
plt.plot(x, x + 5, linestyle=':')
plt.plot(x, x + 6, linestyle='-.')
```

### 调整坐标轴

```python
plt.xlim(-1, 11)  # x 坐标系范围
plt.ylim(-1.5, 1.5)
plt.plot(x, np.sin(x))
```

**一个函数设置 x，y 的刻度的范围**

```python
plt.axis([-1, 11, -1.5, 1.5])
# axis 也可以 收紧坐标系，不留空白
plt.axis('tight')
plt.plot(x, np.cos(x))
```

```python
plt.axis('equal') # x 和 y 的刻度相同一样长
plt.plot(x, np.cos(x))
```

### 图形标签

- title 标题
- xlabel x轴标题
- ylabel
- label 图形线的 标题

**使用matlab方式的画线方法**

- plt.xlabel()  ax.set_xlabel()
- ... 其他方法类似

**ax = plt.axes() 表示 坐标轴**

```python
## set 方法一次设置全部
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2), xlabel='x', ylabel='y', title="set")
```

# 绘制散点图

## 使用plot 画散点图

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, 'o', color='black') # o 是marker 
```

```python
# 多种组合
plt.plot(x, y, '-ok') # k 是黑色， - 是 直线， 
```

## 使用scatter 画散点图

```python
plt.scatter(x, y, marker='o')
```

```python
rng = np.random.RandomState(0)
```

```python
x = rng.randn(100)
y = rng.randn(100) # randn 复合 正态分布
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar() # 显示颜色条
```

**大型数据集 推荐使用 plot 画散点图**

# 基本误差线

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)  # yerr 竖直的 误差线
plt.errorbar(x, y, yerr=dy, fmt='.k')
```

## 密度图和等高线图

**contour外形**

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)  # meshgrid 网格 mesh 网孔， grid网格
Z = np.sin(X)** 10 + np.cos(10 + Y * X) * np.cos(X)

plt.contour(X, Y, Z, colors='black')
```

```python
print(X.shape, Y.shape, Z.shape)
```

```python
plt.contour(X, Y, Z, cmap='RdGy')
plt.colorbar()
```

```python
plt.contourf(X, Y, Z, cmap='RdGy')  # 加个f 填充， 充色的可视化图
```

**使用imshow效果更好**
- imshow 不支持设置 xlim, 而是使用extent 设置

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
```
